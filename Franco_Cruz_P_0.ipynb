## Previo de la Práctica 0

### Paquetes en ROS2



En ROS2, un paquete es una unidad organizativa creada para facilitar la descarga y distribución del código trabajado, siendo así más sencillo que más personas compilen y utilicen este. La creación de un paquete usa **ament** y **colcon** como sistema y herramienta de compilación respectivamente. Un paquete se puede crear utilizando tanto CMake como Python, siendo el segundo el utilizado a lo largo de la práctica, aclarar esto es importante debido a que la estructura de un paquete cambia dependiendo de cual de los dos usamos [1].

Cuando trabajamos con Python, los componentes minimos son:

- **package.xml:** Archivo que contiene metainformación del paquete.

- **resource/< nombre_del_paquete >:** Archivo marcador para el paquete.

- **setup.cfg:** Es requerido cuando el archivo tiene ejecutables, de tal forma que <span style='color:blue'>ros2 run</span> pueda encontrarlo.

- **setup.py:** Contiene instrucciones de como instalar el paquete.

- **< nombre_del_paquete >:** Es un directoria con el mismo nombre que tu paquete, el cual usan las herramientas de ROS2 para encontrar el paquete.

Se pueden tener más de un paquete dentro de un mismo espacio de trabajo, sin importar si se trabaja con Python o CMake, pero cada uno debe de estar en su propia carpeta y no deben de estar anidados. Para estos casos, lo ideal es tener una carpeta **src**, en la cual se crearan los paquetes, obteniendo así un nivel superior del espacio de trabajo limpio [1].

### Nodos y topicos.

Para comprender en que consisten los nodos y topicos es necesario comprender su funcionamiento. Al momento de analizar un sistema, nos encontramos con numerosas variables que requieren de su respectivo control, por lo que abordar todas al mismo tiempo se convierte en una tarea un tanto complicada. Es por esto que ROS2 se encarga de dividir todo el sistema en "subsistemas" encargados de una sola función, estas "subfunciones" son llamadas _nodos_ [2].

Pero el sistema no puede funcionar simplemente con los nodos, si no que es necesaria la comunicación entre estos, necesitando de un puente entre ellos. Los topicos son los "puentes" entre los distintos nodos, pero es necesario tener ciertas consideraciones para poder realizar esta conexión entre nodos. Es aquí donde surgen los tres tipos de nodos que conocemos: el nodo publicador, el nodo subscriptor y el nodo subscriptor-publicador.

El nodo publicador se encarga de enviar datos, el cual pasara por el topico y lo conectará a un nodo subscriptor, el cual se encarga de recibir estos datos para realizar su función, pero hay casos en los que va a ser necesario enviar otro mensaje basado en el recibido, es en estos casos en los que un nodo subscriptor se transforma en uno subscriptor-publicador [3].

Un aspecto importante a tener en cuenta de los topicos es que no es solo una via de comunicación entre dos nodos, si no que puede ser entre multiples, pudiendo obtener información de multiples nodos y pudiendo enviarla de igual forma a un gran numero de nodos susbscriptores. Es así que podemos encontrar topicos que conecte a varios publicadores con un solo subscriptor o visceversa [3].


## Desarrollo de la práctica 

El desarrollo de la práctica se divide en dos partes:

Para la realización de la práctica se requirió de una maquina virtual que contaba con ubuntu, esto con el fin de trabajar en un entorno con linux, donde se puede trabajar desde el kernel.

El primer paso para la realización de la práctica fue el familiarizarse con el uso del kernel, una vez adquirido un poco de práctica se prosiguió con la creación del paquete, para lo cual, nos trasladamos al espacio de trabajo ws_2026-2 y a la carpeta src,donde se creó un paquete con el comando <span style='color:blue'>ros2 pkg create --build-type ament_python</span> (En nuestro caso agregamos también la licencia del MIT).

Una vez ejecutado el comando el paquede debe de haberse creado, una vez dentro de Visual Studio Code, se creó una pestaña nueva la cual corresponde al nodo publicador, donde se realizó la lógica para publicar un número entero, espeficicamente el número "2" constantemente cada segundo estando asociado al topico "topic_1", en el caso del nodo subscriptor-publicador lo programamos para que recibiera la información por medio del topico  "topic_1" y que publicara a travéz del topico "Counter..." la suma acumulada de los números recibidos.

Para poder comprobar el funcionamiento de los códigos fue necesario agregar la licencia en el package y los nombres de los nodos en el setup.py.

Una vez realizado los cambios anteriores es momento de trabajar nuevamente sobre el kernel, en este es necesario compilar el programa por medio del comando <span style='color:blue'>colcon build --symlink-install && source install/setup.bash</span>, donde la ultima parte es necesario antes de activar los nodos.

Una vez compilado, con el comando <span style='color:blue'>ros2 run p0_py "nodo"</span> se activan el nodo "nodo", este paso es necesario para ambos nodos de la práctica. Ya activos los nodos, se utiliza el comando <span style='color:blue'>ros2 topic echo /topico</span>. donde topico es el nombre de los topicos creados en el código. Con esto, se imprime en el kernel la información que pasa por los topicos, comprobando así la conexión entre ambos nodos y el correcto funcionamiento de ambos.

Para los requerimientos de la práctica se requirieron pocos cambios de lo explicado anteriormente, siendo esto simplemente cambiar para todo lo que estaba en función de Int64 a Float64, así como modificar la función callback de ambos nodos, haciendo que en el nodo publicador tuviera un valor inicial de 0.0, y cada vez que llamara la función callback este aumentara a una razón de pi/60, siendo este valor el argumento de una función senoidal a la que se aplicó un valor absoluto,siendo este el valor que publica el nodo.

Para el nodo subscriptor publicador solo se cambió el valor, que en lugar de ser una suma continua de los valores acumulados que leia del topico, ahora simplemente se multiplicaba a una razón de 2 * pi /60, para pasarlo a rad/s.

Además de lo mencionado se hicieron un par de modificaciones en los nombres de los topicos y de los nodos para una mejor presentación.
