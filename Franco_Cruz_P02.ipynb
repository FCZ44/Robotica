<p style=" font-weight: bold;">Simuladores que son compatibles con ROS 2</p>


De acuerdo con la documentación oficial de ROS 2,ROS 2 es compatible con diversos simuladores avanzados que permiten realizar pruebas virtuales de robots utilizando modelos físicos y sensores simulados.

Entre los simuladores compatibles destacan:

- Gazebo
- Webots
- MVSim

Estos simuladores permiten obtener resultados cercanos a los que se presentan en robots reales, debido a que utilizan modelos físicos para representar sensores, actuadores y objetos del entorno.


- Gazebo

GGazebo es uno de los simuladores más utilizados en ROS2 debido a que permite crear entornos tridimensionales complejos y simular sensores como cámaras, LiDAR e IMU. Además, permite probar algoritmos de navegación, percepción y control antes de implementarlos en un robot físico.

Gazebo ofrece un conjunto completo de herramientas que incluye bibliotecas de desarrollo y servicios en la nube para facilitar la creación de simulaciones. Gracias a esto, es posible realizar pruebas en ambientes realistas utilizando modelos físicos y flujos de sensores de alta fidelidad.

Este simulador permite a los desarrolladores iterar rápidamente en nuevos diseños robóticos, evaluar estrategias de control de manera segura y realizar pruebas antes de trabajar con hardware real. Asimismo, Gazebo puede emplearse en procesos de integración continua para verificar el funcionamiento de algoritmos y sistemas robóticos de forma automatizada.

La integración entre ROS2 y Gazebo facilita la comunicación entre nodos, sensores y actuadores mediante topics y servicios, permitiendo ejecutar simulaciones muy similares al comportamiento de un robot real. Sin embargo, no todas las versiones de Gazebo son compatibles con todas las distribuciones de ROS2. La compatibilidad depende de la combinación de versiones instaladas, por lo que es importante verificar la documentación oficial antes de configurar el entorno de simulación. Es el estándar ROS REP-2000 el que define cuál es la versión predeterminada de Gazebo recomendada para cada distribución de ROS2.


- Webots

Webots es un simulador multiplataforma que facilita el diseño y prueba de robots móviles y manipuladores. Cuenta con integración oficial con ROS2, es decir, proporciona una interfaz entre ROS 2 y Webots. Incluye varios subpaquetes, entre ellos webots_ros2_driver, que permite iniciar Webots y comunicarse con él.

- MVSim

MVSim (MultiVehicle Simulator) es un simulador ligero y de código abierto orientado principalmente a robots móviles terrestres. Se utiliza para realizar pruebas de navegación, mapeo y control cinemático en entornos virtuales antes de implementar algoritmos en robots reales. MVSim emplea simulación física en 2D con visualización en 3D, permitiendo representar distintos tipos de vehículos, como robots de tracción diferencial y vehículos Ackermann. Además, soporta múltiples sensores como LiDAR, cámaras, IMU y GPS, así como integración nativa con ROS2 mediante mensajes estándar para la comunicación entre nodos.

El simulador incluye diferentes mundos de demostración que permiten probar funcionalidades como simulación multirobot, configuraciones de sensores, distintos tipos de terreno y vehículos articulados. Estos escenarios pueden ejecutarse como una aplicación independiente utilizando la herramienta mvsim o como un nodo de ROS 2 capaz de publicar datos de sensores y recibir comandos de velocidad mediante topics estándar. Los entornos de simulación se definen mediante archivos XML con extensión .world.xml, en los cuales se describen los vehículos, obstáculos, sensores y parámetros físicos de la simulación. Además, MVSim proporciona bibliotecas de vehículos y sensores predefinidos que pueden reutilizarse y modificarse para crear entornos personalizados.

- Otros simuladors compatibles

Además de Gazebo, Webots y MVSim, ROS2 también es compatible con otros simuladores utilizados en investigación y desarrollo robótico, como CoppeliaSim, usado para simulación de robots móviles y manipuladores; NVIDIA IsaacSim, enfocado en inteligencia artificial y visión artificial; o también CARLA y AirSim, que son utilizados principalmente en el desarrollo de vehículos autónomos y drones.


<p style=" font-weight: bold;">Investigar cuales son las restricciones del simulador Gazebo 2</p>


Gazebo presenta restricciones de compatibilidad entre sus diferentes versiones y las distribuciones de ROS 2, no todas las versiones de Gazebo son compatibles con todas las distribuciones de ROS 2. El estándar ROS REP-2000 establece cuál es la versión predeterminada de Gazebo que debe utilizarse para cada distribución de ROS 2, con el objetivo de asegurar un funcionamiento estable y compatible entre paquetes, plugins y herramientas de simulación.  Además, algunas funcionalidades dependen de plugins adicionales para sensores y actuadores, los cuales pueden variar según la versión instalada.

Otra restricción importante es el alto consumo de recursos computacionales. Las simulaciones que incluyen múltiples robots, sensores avanzados o entornos complejos pueden requerir procesadores multinúcleo y tarjetas gráficas dedicadas para mantener un desempeño adecuado. Además, aunque Gazebo ofrece motores físicos avanzados, la simulación nunca puede reproducir completamente las condiciones del mundo real. Fenómenos como la deformación de materiales, la fricción no lineal, los contactos complejos entre objetos o ciertos comportamientos de agarre son difíciles de modelar con total precisión. De igual forma, el realismo visual del entorno, aunque suficiente para muchas aplicaciones de robótica, puede resultar limitado en comparación con simuladores orientados al entrenamiento de inteligencia artificial basada exclusivamente en visión computacional.

Algunos dispositivos especializados, como cámaras térmicas o determinados sensores LiDAR, pueden necesitar configuraciones manuales complejas e incluso modificaciones en el código para lograr un comportamiento realista. En consecuencia, aunque Gazebo constituye una herramienta muy poderosa para el diseño, prueba y validación de sistemas robóticos, no representa una solución universal para todos los problemas. La elección de utilizar Gazebo dependerá de los objetivos específicos del proyecto, siendo una excelente opción para simulaciones robóticas integradas con ROS 2, pero pudiendo resultar menos adecuada en aplicaciones que requieran un alto nivel de realismo visual o una reproducción exacta de todos los fenómenos físicos del mundo real.

<p style=" font-weight: bold;">1. Explicar las relaciones de los archivos xacros.</p>

Los archivos .xacro son archivos que nos permiten generar la descripción física de los modelos que queremos implementar. En nuestro caso, para generar un modelo SCARA funcional con dos sensores, contamos con 5 descripciones en archivos xacro de los componentes del modelo:

- scara_urdf.xacro: de una construcción urdf (Unified Robot Description Format), generamos un archivo .xacro que nos permite construir los tres links del robot base, con sus respectivos joints entre links, además de un link base, un word link y un link p, con sus respetivos joints. Lo que genera este archivo .xacro es la base del modelo.

- Archivos de sensores, camera_sensor.xacro y lidar_sensor.xacro: en cada uno de ellos, se construye geométricamente cada sensor y se describe su operación, de tal forma que el modelo completo de nuestro robot tiene una cámara como sensor y un sensor de intensidad de luz, cada uno de ellos unido al link p definido en el modelo scara_urdf.xacro, por lo que al moverse el robot se mueven todos sus sensores junto con él.

- scara_gz2.xacro: a través de una configuración de control PID, se logra mantener estables a los movimientos de las articulaciones del robot. Este control permite que el modelo en gazebo se comporte adecuadamente.

- gz2_scara.xacro: este archivo xacro nos permite monter todos los archivos individuales .xacro como un conjunto. Dentro de su estructura no se cuenta con descripciones físicas de los elementos del robot y sus partes, más bien, se aprovecha que cada parte ya se construyó y se ensambla. Gracias a este archivo, todo lo que se cosntruyó indivualmente puede conectarse como un conjunto.


<p style=" font-weight: bold;">2. Decribir las partes del archivo xacro de la configuración de las juntas del robot.</p>


```python
    <joint name="world_joint" type="fixed">
        <parent link="world"/>
        <child link="base_link"/>
        <origin xyz="0 0 0" rpy="0 0 0"/>
    </joint>
```
    *Sección de código que genera un joint en el ensamble del robot.*

   
Todos los joints tienen el formato mostrado. La configuración de una junta comienza con la definición de un nombre y su tipo. A cada junta se le asocian dos links, un link parent que define donde inicia la junta y le dicta cómo debe moverse, y un link child, al cuál la junta une al link parent y hace que su movimiento esté influenciado por los que realice. Al igual que todos los objetos que se definen en un .xacro, los joints se definen con una posición y y una orientación. Los joints son fundamentales para crear una estructura sólida en el robot y permitir un comportamiento óptimo. Cada junta es una articulación que soporta y une toda la estructura. De igual forma, dejar joints o juntas específicas permite que podamos montar los sensores o cualquier otro instrumento que queramos sobre la estructura del robot.

De forma adicional, las juntas de la estructura base del robot se configuran con un control específico para que su comportamiento no sea errático y se apegue a una estabilidad deseada.

```python
   <gazebo>
        <plugin 
            filename="gz-sim-joint-position-controller-system" 
            name="gz::sim::systems::JointPositionController">
            <joint_name>link_2_joint</joint_name>
            <p_gain>100.0</p_gain>
            <d_gain>30.0</d_gain>
            <i_gain>0.01</i_gain>
            <i_max>1.0</i_max>
            <i_min>-1.0</i_min>
        </plugin>
    </gazebo>
```
    *Sección de código que configura el comportamiento de un joint.*

   
Gracias a este código se configura un controlador PID que, a través de un plugin, el cual le indica al motor de Gazebo que debe realizar una acción. Las gananciad p, d e i generan un control suave en el movimiento de la junta.

<p style=" font-weight: bold;">3. Describir las relaciones del archivo xacro del sensor de la cámara.</p>   


Para configurar la descripción completa del sensor de la cámara a través de un archivo .xacro, se comienza con la construcción geométrica del sensor. A través la construcción de una caja con su respectivas coordenadas de posición y de orientación, además de especificaciones para la masa y la inercia, se obtiene la representación del sensor y los datos generales para que su movimiento pueda calcularlo el simulador.

Posterior a la configuración del modelo del sensor se escriben las lineas sobre las juntas y los eslabones que componen al sistema y que son de suma importancia para que el simulador sepa donde debe ir montado este módulo extra. 

Seguido de esta sección del código se encuentra la sección de configuración del comportamiento del sensor. 

```python
  <gazebo reference="camera_link">
        <sensor name="camera" type="camera">
            <pose>0 0 0 0 0 0</pose>
            <visualize>true</visualize>
            <update_rate>30.0</update_rate>
            <camera>
                <horizontal_fov>1.089</horizontal_fov>
                <image>
                    <width>640</width>
                    <height>480</height>
                    <format>R8G8B8</format>
                </image>
                <clip>
                    <near>0.05</near>
                    <far>8.0</far>
                </clip>
            </camera>
            <topic>camera/image_raw</topic>
            <gz_frame_id>camera_link_optical</gz_frame_id>
        </sensor>
    </gazebo>
```
A través de este código se le indica al simulador gazebo que para el link llamado camera_link debe crear un sensor de tipo camara. Las siguientes lineas habla de la orientación relativa del nuevo sensor, de la visualización que tendrá en Gazebo y se establece una frecuencia de actualización de 30 FPS. 

La siguiente sección del código configura los parámetros ópticos de la cámara. Se tiene 1.089 radianes de amplitud de visión de la cámara, una imagen de 640x480 pixeles de resolución y un formato RGB de 24 bits, además se establecen los límites de visión de la camara, por lo que en este modelo en particular se habla de un intervalo de 0.5 a 8 metros de visión de la cámara.
   
Por último, se establece cómo el sensor transmitirá su información al sistema de tópicos del robot, por lo que se le indica que las imágenes deben ir al topico camera/image_raw, con la orientaci{on establecidad gracias al sistema de referencia camera_link_optical.

- launch

El sistema de lanzamiento de ROS 2 (launch) es una herramienta que sirve para organizar y ejecutar todos los elementos necesarios en una aplicación robótica de manera automática. Normalmente se utilizan varios nodos al mismo tiempo, por ejemplo nodos de control, sensores, simulación y visualización, por lo que iniciar cada uno manualmente es complicado y tardado. Por esta razón, ROS 2 utiliza archivos launch, los cuales permiten definir qué programas deben ejecutarse, con qué parámetros trabajarán y cómo se comunicarán entre sí. Además, este sistema ayuda a reutilizar componentes, ya que un mismo nodo puede funcionar de manera distinta dependiendo de la configuración que se le asigne en el archivo launch.

Otra característica importante es que el sistema de lanzamiento también supervisa los procesos que se encuentran activos. Esto significa que puede detectar si un nodo presenta errores, se detiene inesperadamente o cambia de estado, permitiendo informar al usuario o ejecutar acciones automáticas para mantener el funcionamiento del sistema. Los archivos launch pueden escribirse en distintos formatos, como Python, XML o YAML, aunque Python es uno de los más utilizados debido a su flexibilidad. Dentro de ROS 2, el paquete encargado de manejar esta funcionalidad es launch_ros, el cual trabaja junto con la infraestructura general llamada launch. Todo este sistema fue diseñado para facilitar el manejo de aplicaciones robóticas complejas, haciendo que la ejecución y coordinación de múltiples procesos sea más ordenada, flexible y sencilla para el usuario.

El archivo gz2_scara.launch.py es el código principal de lanzamiento del paquete scara_bringup y su función es iniciar automáticamente todo el entorno de simulación del robot SCARA en Gazebo utilizando ROS 2. Después de iniciar Gazebo inserta el robot SCARA definido en el paquete scara_description, iniciar la comunicación entre Gazebo y ROS 2, y activar herramientas de visualización como, en este caso, RViz. Este archivo launch permite que  todo el entorno de simulación pueda ejecutarse mediante un solo comando en la terminal, facilitando el trabajo y la organización del sistema robótico.

- Explicación del código gz2_scara.launch.py
```python
import os
from ament_index_python.packages import get_package_share_directory
from launch import LaunchDescription
from launch.actions import ExecuteProcess
from launch_ros.actions import Node
from launch.substitutions import Command

def generate_launch_description():

    pkg_scara_description = get_package_share_directory('scara_description')
    pkg_scara_bringup = get_package_share_directory('scara_bringup')

    urdf_path = os.path.join(pkg_scara_description, 'urdf', 'gz2_scara.xacro')
    gazebo_config_path = os.path.join(pkg_scara_bringup, 'config', 'gz_bridge.yaml')
    world_path = os.path.join(pkg_scara_bringup, 'worlds', 'world_config.world')

    rviz_config_path = os.path.join(pkg_scara_bringup, 'rviz', 'scara_config.rviz')

    use_sim_time = {'use_sim_time': True}

    robot_state_publisher_node = Node(
        package='robot_state_publisher',
        executable='robot_state_publisher',
        parameters=[{'robot_description': Command(['xacro ', urdf_path])}, use_sim_time]
    )

    gz_ros_bridge_node = Node(
        package='ros_gz_bridge',
        executable='parameter_bridge',
        parameters=[{'config_file': gazebo_config_path}, use_sim_time],
        output='screen'
    )

    gz_sim = ExecuteProcess(
        cmd=['gz', 'sim', '-r', world_path, '--render-engine', 'ogre'],
        output='screen'
    )

    spaw_entity = Node(
    package='ros_gz_sim',
    executable='create',
    arguments=[
        '-topic', 'robot_description',
        '-name', 'robot_scara',
        '-world', 'world_config' # DEBE coincidir con el <world name="..."> de tu archivo .world
    ],
    output='screen'
    )

    rviz_node = Node(
        package='rviz2',
        executable='rviz2',
        name='rviz2',
        output='screen',
        arguments=['-d', rviz_config_path],
        parameters=[use_sim_time] # Fundamental para evitar errores de TF
    )

    return LaunchDescription([
        robot_state_publisher_node,
        gz_sim,
        spaw_entity,
        gz_ros_bridge_node,
        rviz_node
    ])
```

Se hace la importación necesaria de diferentes librerías para crear el launch, ejecutar procesos y lanzar nodos dentro de ROS 2. Después, mediante la función get_package_share_directory, se obtienen las rutas de los paquetes scara_description y scara_bringup. 

Posteriormente, se define la ruta del archivo gz2_scara.xacro, que contiene la descripción del robot SCARA en formato Xacro/URDF. También se define el archivo gz_bridge.yaml, descrito anteriormente, y el archivo world_config.world, que contiene el escenario virtual donde se ejecutará la simulación. Además, se carga un archivo de configuración de RViz para visualizar  el robot y sus transformaciones.

Se activa el parámetro use_sim_time, lo que permite que todos los nodos utilicen el tiempo de simulación de Gazebo en lugar del tiempo real del sistema. 

El launch crea varios nodos importantes para el funcionamiento del sistema:

    - robot_state_publisher_node
    - gz_ros_bridge_node
    - spaw_entity
    - rviz_node

El nodo robot_state_publisher publica la descripción y el estado de las articulaciones del robot utilizando el archivo Xacro convertido dinámicamente a URDF mediante el comando xacro. Después se inicia el nodo gz_ros_bridge, el cual crea la comunicación entre ROS 2 y Gazebo utilizando la configuración definida en gz_bridge.yaml.

Posteriormente, gz_sim corresponde a un proceso ejecutado mediante ExecuteProcess, cuya función es iniciar Gazebo y cargar el mundo de simulación definido en el archivo .world. Este proceso no corresponde a un nodo ROS 2, sino a la ejecución externa del simulador Gazebo desde el archivo launch.

El nodo spaw_entity, del paquete ros_gaz_sim. tiene la función de crear e insertar el robot SCARA dentro del entorno de simulación de Gazebo. Para ello, obtiene la descripción del robot desde el tópico robot_description, donde previamente se publica el modelo definido en el archivo Xacro/URDF. Posteriormente, el nodo genera el robot virtual y lo coloca dentro del mundo de simulación llamado world_config con el nombre robot_scara.

Finalmente, se inicia RViz con el archivo de configuración scara_config.rviz, permitiendo visualizar el robot, sus transformaciones TF y otros elementos gráficos de la simulación.

En conjunto, este archivo launch automatiza completamente el entorno de simulación del robot SCARA, ya que con un solo comando se ejecutan Gazebo, la comunicación con ROS 2, la carga del robot y la visualización en RViz, facilitando el desarrollo y prueba de aplicaciones robóticas.

El comando utilizado en terminal para ejecutar todo lo descrito anteriormente es: 

```python
    ros2 launch scara_bringup gz2_scara.launch.py
```

<p style=" font-weight: bold;">5. Describir que hace el archivo bridge. </p>

El archivo bridge se ejecuta a través de parameter_bridge del paquete ros_gz_bridge, su función es la de un puente de red bidireccional en tiempo real. En el código del launch el ejecutable parameter_bridge lee la configuración almacenada en gz_bridge.yaml y crea automáticamente los puentes de comunicación especificados. Conecta dos ecosistemas de robótica: ROS 2 y Gazebo.

El archivo bridge resulta fundamental en ROS2 porque de lo contrario la comunicación sufre fallas, principalmente por dos razones: 

- Los protocolos de red son distintos (Middlewares): ROS 2 utiliza DDS (Data Distribution Service) para enviar datos por la red, mientras que Gazebo utiliza Gazebo Transport, basado en sockets ZeroM.

- Los formatos de datos son incompatibles (Serialización): ROS 2 empaqueta la información (.msg) estructurada bajo el estándar CDR. Gazebo empaqueta su información utilizando Google Protocol Buffers (Protobuf).


Las funciones del archivo bridge son esencialmente las siguientes:

  - Enviar comandos de movimiento desde ROS 2 hacia Gazebo.
  - Recibir datos de sensores simulados en Gazebo y publicarlos en ROS 2.
  - Compartir información de cámaras, LIDAR, IMU o estados de articulaciones.
  - Sincronizar el tiempo de simulación (/clock) entre ambos sistemas.

- Explicación del código gz_bridge.yaml
```python
- ros_topic_name: "/clock"
  gz_topic_name: "/clock"
  ros_type_name: "rosgraph_msgs/msg/Clock"
  gz_type_name: "gz.msgs.Clock"
  direction: GZ_TO_ROS

- ros_topic_name: "/joint_states"
  gz_topic_name: "/world/world_config/model/robot_scara/joint_state"
  ros_type_name: "sensor_msgs/msg/JointState"
  gz_type_name: "gz.msgs.Model"
  direction: GZ_TO_ROS

- ros_topic_name: "/camera/image_raw"
  gz_topic_name: "/camera/image_raw"
  ros_type_name: "sensor_msgs/msg/Image"
  gz_type_name: "gz.msgs.Image"
  direction: GZ_TO_ROS

- ros_topic_name: "/tf"
  gz_topic_name: "/model/scara/tf"
  ros_type_name: "tf2_msgs/msg/TFMessage"
  gz_type_name: "gz.msgs.Pose_V"
  direction: GZ_TO_ROS

- topic: "/world/world_config/model/robot_scara/joint_state"
  gz_type: "gz.msgs.Model"
  ros_type: "sensor_msgs/msg/JointState"
  direction: GZ_TO_ROS

- ros_topic_name: "/joint1/cmd_pos"
  gz_topic_name: "/model/robot_scara/joint/link_1_joint/0/cmd_pos"
  ros_type_name: "std_msgs/msg/Float64"
  gz_type_name: "gz.msgs.Double"
  direction: ROS_TO_GZ

- ros_topic_name: "/joint2/cmd_pos"
  gz_topic_name: "/model/robot_scara/joint/link_2_joint/0/cmd_pos"
  ros_type_name: "std_msgs/msg/Float64"
  gz_type_name: "gz.msgs.Double"
  direction: ROS_TO_GZ

- ros_topic_name: "/joint3/cmd_pos"
  gz_topic_name: "/model/robot_scara/joint/link_3_joint/0/cmd_pos"
  ros_type_name: "std_msgs/msg/Float64"
  gz_type_name: "gz.msgs.Double"
  direction: ROS_TO_GZ

- ros_topic_name: "/scan"
  gz_topic_name: "/scan"
  ros_type_name: "sensor_msgs/msg/LaserScan"
  gz_type_name: "gz.msgs.LaserScan"
  direction: GZ_TO_ROS
```

El archivo gz_bridge.yaml indica qué tópicos deben conectarse entre ROS2 y Gazebo, qué tipo de mensajes utilizan y en qué dirección debe realizarse la comunicación. Sin este archivo, los nodos de ROS2 no podrían intercambiar información con la simulación de Gazebo, por lo que el robot virtual no podría ser controlado ni monitoreado desde ROS2, el robot SCARA se quedaría estático en el mundo virtual. El código de control enviaría comandos al vacío, y las gráficas de RViz se quedarían congeladas por no recibir la posición real del robot.

En el archivo .yaml el tópico /clock transmite el tiempo de simulación de Gazebo hacia ROS2 para que todos los nodos trabajen sincronizados con el tiempo simulado. También se publica el estado de las articulaciones del robot mediante el tópico /joint_states, permitiendo que ROS 2 conozca en todo momento la posición de cada articulación del SCARA.

Por otra parte, el bridge permite enviar a ROS2 la información generada por sensores simulados. El tópico /camera/image_raw transmite las imágenes captadas por la cámara virtual, mientras que el tópico /scan envía las mediciones del sensor láser tipo LiDAR mediante mensajes LaserScan. De la misma forma, se establece un puente para el tópico /tf, que contiene las transformaciones espaciales entre los diferentes enlaces del robot, indispensables para la visualización y el cálculo de posiciones dentro del sistema.

Nótese que existen tres puentes configurados con dirección ROS_TO_GZ, comunicación de ROS a gazebo, estos tres puentes corresponden a los tópicos /joint1/cmd_pos, /joint2/cmd_pos y /joint3/cmd_pos. Estos permiten que ROS2 envíe comandos de posición a las tres articulaciones del robot SCARA dentro de Gazebo. En el momento en que un nodo publica un valor en alguno de estos 3 tópicos, el bridge traduce el mensaje y lo envía al simulador, provocando el movimiento de la articulación correspondiente.

En conjunto, el archivo gz_bridge.yaml actúa como un traductor entre ROS2 y Gazebo, permitiendo que ROS2 reciba información del robot y de los sensores simulados, mientras que Gazebo puede recibir comandos de movimiento para controlar las articulaciones del SCARA. Gracias a este puente bidireccional, es posible supervisar y controlar el robot virtual desde aplicaciones desarrolladas en ROS2.
